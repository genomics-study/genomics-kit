```python
df_phen, df_xp = read_dfs(nan_fill=0)
```

```python
y = df_phen['"grupa"'].values
y[(y == 1) | (y == 3)] = 0
y[(y == 2) | (y == 5)] = 1
```

```python
without_4 = (y != 4)
```

```python
X = df_phen.drop('"grupa"', axis=1).values
```

```python
X_l = np.concatenate((X[6:23, :][without_4[6:23]], X[31:43, :][without_4[31:43]], X[48:, :][without_4[48:]])) 
y_l = np.concatenate((y[6:23][without_4[6:23]], y[31:43][without_4[31:43]], y[48:][without_4[48:]])) 

X_t = np.concatenate((X[:6, :][without_4[:6]], X[23:31, :][without_4[23:31]], X[43:48, :][without_4[43:48]]))
y_t = np.concatenate((y[:6][without_4[:6]], y[23:31][without_4[23:31]], y[43:48][without_4[43:48]])) 
```

```python
model.fit(X_l, y_l)
```

```python
model.set_validation(Validator())
model.validate(X_t, y_t)
```
